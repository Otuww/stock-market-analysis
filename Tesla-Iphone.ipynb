```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Data olish
apple = yf.download("AAPL", start="2022-01-01", end="2023-01-01")
tesla = yf.download("TSLA", start="2022-01-01", end="2023-01-01")

# Close narxlarni olish
apple_close = apple['Close']
tesla_close = tesla['Close']

# Grafik
plt.figure(figsize=(10,5))
plt.plot(apple_close, label="Apple")
plt.plot(tesla_close, label="Tesla")

plt.title("Apple vs Tesla Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# O'rtacha narxlar
print("Apple average:", apple_close.mean())
print("Tesla average:", tesla_close.mean())
```
